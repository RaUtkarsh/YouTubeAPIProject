## YouTube API Project EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('YouTubeAPIprojectDataset.csv')
df
```

```python
df.info()
```

```python
df['durationInMin'] = df['duration'].div(60)
df['durationInMin']
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (15, 13))

# Create the first boxplot in the first row and first column
sns.boxplot(data=df, y='duration', ax=ax[0][0])

# Create the second boxplot in the first row and second column
sns.boxplot(data=df, y='viewCount', ax=ax[0][1])

# Create the third boxplot in the second row and first column
sns.boxplot(data=df, y='likeCount', ax=ax[1][0])

# Create the fourth boxplot in the second row and second column
sns.boxplot(data=df, y='commentCount', ax=ax[1][1])

# Show the plot
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['figure.facecolor'] = 'white'

df_new = df.copy()
df_new = df_new.sort_values('viewCount', ascending=False)
df_new = df_new.head(5)

sns.barplot(data = df_new, x = 'title', y = 'viewCount')

plt.title('Most Viewed Videos')
plt.xlabel("Video Title")
plt.ylabel("Views")
plt.xticks(rotation = 90)
plt.show()

del df_new
```

```python
df_copy = df.copy()
df_copy = df_copy.sort_values('likeCount', ascending = False)
df_copy = df_copy.head(5)

sns.barplot(data = df_copy, x = 'title', y = 'likeCount')
plt.title('Most Liked Videos')
plt.xlabel('Video Title')
plt.ylabel('Likes')
plt.xticks(rotation = 90)

plt.show()
```

```python
cnt_videos = df.groupby('publishedDay').size().sort_values(ascending = False)
plt.bar(cnt_videos.index, cnt_videos.values)
plt.title('Number of videos uploaded on each day of the week')
plt.xticks(cnt_videos.index)
plt.show()
```
